```python
df = genfromtxt('housing_data.csv',delimiter=',')
qw = genfromtxt('housing_data.csv',delimiter=',')
last=df[:,-1]
print(last.shape)
medu=last
#print(medu)
feature=df[:,0:13]
#print(feature)
hou=qw[:,-1]
qian=qw[:,0:13]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
def linear_regression(model,dataset, allpre_res): 
    train,test = [],[]
    X_train, X_test, y_train, y_test = train_test_split(dataset, medu, test_size=0.9, random_state=101)
    model.fit(X_train, y_train)
    train_pred_y = model.predict(X_train)
    test_pred_y = model.predict(X_test)
    train.append(mean_squared_error(y_train, train_pred_y))
    test.append(mean_squared_error(y_test, test_pred_y))
    print("training RMSE: %lf" %sqrt(np.mean(train)), "testing RMSE: %lf" %sqrt(np.mean(test)))

    return sqrt(np.mean(train)), sqrt(np.mean(test)) 
```

```python
def plotScatter(Y_fitted,Y,regressor_name):
    index=[i for i in range(1,507)]
    index=np.asarray(index)
    Y_plot=Y[0:506]
    Y_fitted_plot=Y_fitted[0:506]
    plt.figure(figsize=(10,8))
    plt.scatter(index,Y_plot,c='b',label='True value')
    plt.scatter(index,Y_fitted_plot,c='r',label='Fitted value')
    plt.title('fitted vs true value for '+regressor_name)
    plt.legend(loc=1,ncol=1, borderaxespad=0.)
    plt.show()

    Y_residual=Y_plot-Y_fitted_plot
    plt.figure(figsize=(10,8))
    plt.scatter(index,Y_residual,c='g',label='Residual')
    plt.scatter(index,Y_fitted_plot,c='y',label='Fitted value')
    plt.title('Residual vs Fitted value for '+regressor_name)
    plt.legend(loc=1,ncol=1, borderaxespad=0.)
    plt.show()
```

```python
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_validate
from sklearn.tree import export_graphviz
import warnings
import os
from PIL import Image
from sklearn.preprocessing import OneHotEncoder
from sklearn.neural_network import MLPRegressor
data3= pd.read_csv('insurance_data.csv')
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
def plotScatter2(Y_fitted,Y,data_number,regressor_name):
    index=[i for i in range(1,data_number+1)]
    index=np.asarray(index)
    Y_plot=Y[0:data_number]
    Y_fitted_plot=Y_fitted[0:data_number]
    plt.figure(figsize=(10,8))
    plt.scatter(index,Y_plot,c='b',label='True value')
    plt.scatter(index,Y_fitted_plot,c='r',label='Fitted value')
    plt.title('fitted vs true value for '+regressor_name)
    plt.legend(loc=1,ncol=1, borderaxespad=0.)
    plt.show()

    Y_residual=Y_plot-Y_fitted_plot
    plt.figure(figsize=(10,8))
    plt.scatter(index,Y_residual,c='g',label='Residual')
    plt.scatter(index,Y_fitted_plot,c='y',label='Fitted value')
    plt.title('Residual vs Fitted value for '+regressor_name)
    plt.legend(loc=1,ncol=1, borderaxespad=0.)
    plt.show()
```

```python
def printRMSE(X,Y,model):
    train,test = [],[]
    foldnumber = KFold(n_splits = 10, random_state = None, shuffle = False)
    #regr = linear_model.LinearRegression()
    for train_index, test_index in foldnumber.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = Y[train_index], Y[test_index]
        model.fit(X_train, y_train)
        train_pred_y = model.predict(X_train)
        test_pred_y = model.predict(X_test)
        train.append(mean_squared_error(y_train, train_pred_y))
        test.append(mean_squared_error(y_test, test_pred_y))
    print("training RMSE: %lf" %np.sqrt(np.mean(train)), "testing RMSE: %lf" %np.sqrt(np.mean(test)))
```

## compare effect of different feature processing combinations

```python
data3_enc=data3
le=preprocessing.LabelEncoder()
data3_enc.loc[:,'ft4']=le.fit_transform(data3_enc.loc[:,'ft4'])
data3_enc.loc[:,'ft5']=le.fit_transform(data3_enc.loc[:,'ft5'])
data3_enc.loc[:,'ft6']=le.fit_transform(data3_enc.loc[:,'ft6'])


data3_456=enc.fit_transform(data3_enc.loc[:,'ft4':'ft6']).toarray()
data3_123=data3_enc.loc[:,'ft1':'ft3'].values
data3_trans=np.concatenate((data3_123,data3_456),axis=1)
X3=data3_trans
Y3=np.asarray(data3['charges'].values)

Linreg=LinearRegression()
printRMSE(X3,Y3,Linreg)
Linreg.fit(X3,Y3)
Y3_fitted=Linreg.predict(X3)
RMSE_whole=np.sqrt(mean_squared_error(Y3, Y3_fitted))
print('whole dataset without target transformed',RMSE_whole)
plotScatter2(Y3_fitted,Y3,Y3.shape[0],'Linear regression with One hot encoding only')
```

```python
scalar=StandardScaler()
data3_123_2=scalar.fit_transform(data3_123)
#print(data3_123_2)
data3_trans2=np.concatenate((data3_123_2,data3_456),axis=1)
X3_2=data3_trans2
Linreg2=LinearRegression()
printRMSE(X3_2,Y3,Linreg2)

Linreg2.fit(X3_2,Y3)
Y3_fitted_2=Linreg2.predict(X3_2)
plotScatter2(Y3_fitted_2,Y3,Y3.shape[0],'Linear regression with standardization')
```

```python
data3_1=data3_enc.loc[:,'ft1'].values
for i in range(len(data3_1)):
    if data3_1[i] < 30 and data3_1[i]>=0:
        data3_1[i]=1
    elif data3_1[i]>= 30 and data3_1[i]<=50:
        data3_1[i]=2
    elif data3_1[i]>50:
        data3_1[i]=3
```

```python
data3_1_exp= np.expand_dims(data3_1, axis=1)#add a dimension to data3_1 for dimension alignment 
data3_23=data3_enc.loc[:,'ft2':'ft3'].values
data3_23=scalar.fit_transform(data3_23)
data3_trans3=np.concatenate((data3_1_exp,data3_23,data3_456),axis=1)
X3_3=data3_trans3
Linreg3=LinearRegression()
printRMSE(X3_3,Y3,Linreg3)

Linreg3.fit(X3_3,Y3)
Y3_fitted_3=Linreg2.predict(X3_3)
RMSEtemp=np.sqrt(mean_squared_error(Y3_fitted_3, Y3))
plotScatter2(Y3_fitted_3,Y3,Y3.shape[0],'Linear regression with range coding')
```

```python
# Polynomial parameter tuning :order 3 performance is worse than 2
poly=PolynomialFeatures(3)
X4=poly.fit_transform(X3)
Linreg=LinearRegression()
printRMSE(X4,Y3,Linreg)
Linreg.fit(X4,Y3)
Y3_fitted=Linreg.predict(X4)
RMSE_whole=np.sqrt(mean_squared_error(Y3, Y3_fitted))
X4.shape
```

```python
# random forest 
poly=PolynomialFeatures(2)
X4=poly.fit_transform(X3)
regr2=RandomForestRegressor(n_estimators=20,max_depth=4,max_features=5,bootstrap=True,oob_score=True)
regr2.fit(X4,Y3)
printRMSE(X4,Y3,regr2)
Y3_fitted=regr2.predict(X4)
RMSE_whole=np.sqrt(mean_squared_error(Y3, Y3_fitted))
X4.shape
```

```python
#random forest tuning
poly=PolynomialFeatures(2)
X4=poly.fit_transform(X3)
regr2=RandomForestRegressor(n_estimators=20,max_depth=6,max_features=20,bootstrap=True,oob_score=True)
regr2.fit(X4,Y3)
printRMSE(X4,Y3,regr2)
Y3_fitted=regr2.predict(X4)
RMSE_whole=np.sqrt(mean_squared_error(Y3, Y3_fitted))
X4.shape
```

```python
# Hyperparameter search using Grid CV Random Forest
param_grid={'max_depth': [4,5,6],
            'max_features': [5,10,15],
            'min_samples_leaf': [5, 10, 15],
            'min_samples_split': [5, 10, 15],
            'n_estimators': [20, 30, 40]
           }
# Create a based model
rf = RandomForestRegressor()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 10, n_jobs = -1, verbose = 2)
grid_search.fit(X4, Y3)
printRMSE(X4,Y3,grid_search.best_estimator_)
grid_search.best_params_
```

```python
# neural network
poly=PolynomialFeatures(2)
X4=poly.fit_transform(X3)
neural_regressor = MLPRegressor(hidden_layer_sizes=600,activation='relu',max_iter=300)
neural_regressor.fit(X4,Y3)
printRMSE(X4,Y3,neural_regressor)
Y3_fitted=regr2.predict(X4)
RMSE_whole=np.sqrt(mean_squared_error(Y3, Y3_fitted))
```

```python
# Hyperparameter search using Grid CV Random Forest
neural_param_grid={'hidden_layer_sizes': [200,300,400],
            'learning_rate_init':[0.01,0.005,0.001],
            'activation':['relu','tanh'],
            'max_iter':[200,300,400]
           }
# Create a based model
neural_rf = MLPRegressor()
# Instantiate the grid search model
neural_grid_search= GridSearchCV(estimator = neural_rf, param_grid = neural_param_grid, 
                          cv = 10,n_jobs=-1,verbose=2)
neural_grid_search.fit(X4, Y3)
printRMSE(X4,Y3,neural_grid_search.best_estimator_)
neural_grid_search.best_params_
```

```python
# Gradient boosting
from sklearn import ensemble 

poly=PolynomialFeatures(2)
X4=poly.fit_transform(X3)
params = {'n_estimators': 380, 'max_depth': 4, 'min_samples_split':10,'min_samples_leaf':10,
          'learning_rate': 0.01, 'loss': 'ls'}
boosting_regr = ensemble.GradientBoostingRegressor(**params)
boosting_regr.fit(X4,Y3)
printRMSE(X4,Y3,boosting_regr)
Y3_fitted=boosting_regr.predict(X4)
RMSE_whole=np.sqrt(mean_squared_error(Y3, Y3_fitted))
```
