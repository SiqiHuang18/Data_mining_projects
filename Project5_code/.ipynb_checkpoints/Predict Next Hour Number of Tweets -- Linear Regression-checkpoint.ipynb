```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import datetime, time
import pytz
```

```python
files2 = ['ECE219_tweet_data/tweets_#gohawks.txt', 
          'ECE219_tweet_data/tweets_#gopatriots.txt', 
          'ECE219_tweet_data/tweets_#nfl.txt', 
          'ECE219_tweet_data/tweets_#patriots.txt', 
          'ECE219_tweet_data/tweets_#sb49.txt', 
          'ECE219_tweet_data/tweets_#superbowl.txt']

def getcsv2(file):
    timestamp = []
    tweet_count = []
    followers_count = []
    retweet_count = []
    url_count=[]
    user_mentions=[]
    favorited_num=[]
    ranking_score=[]
    hashtag_number=[]
    with open(file, 'r') as fl:
        for line in fl:
            data = json.loads(line)
            timestamp.append(data['citation_date'])
            tweet_count.append(1)
            followers_count.append(data['author']['followers'])
            retweet_count.append(data['metrics']['citations']['total'])
            url_count.append(len(data['tweet']['entities']['urls']))
            user_mentions.append(len(data['tweet']['entities']['user_mentions']))
            favorited_num.append(data['tweet']['favorite_count'])
            ranking_score.append(data['metrics']['ranking_score'])
            hashtag_number.append(data['title'].count('#'))
        df = pd.DataFrame({
            'tweet' : tweet_count,
            'timestamp' : timestamp,
            'followers' : followers_count,
            'retweeted times' : retweet_count,
            'url count':url_count,
            'user mention':user_mentions,
            'favorited num':favorited_num,
            'ranking score': ranking_score,
            'hashtag count':hashtag_number
        }, columns = ['tweet', 'timestamp', 'followers', 'retweeted times','url count','user mention','favorited num','ranking score','hashtag count'])
        df.to_csv('part1_3'+file[18:-4]+'.csv', index = False)

for file in files2:
    getcsv2(file)
```

```python
newfiles2=['part1_3tweets_#gohawks.csv',
          'part1_3tweets_#gopatriots.csv',
          'part1_3tweets_#nfl.csv',
          'part1_3tweets_#patriots.csv',
          'part1_3tweets_#sb49.csv',
          'part1_3tweets_#superbowl.csv']
```

```python
def preprocess2(file):
    data = pd.read_csv(file)
    data.columns = ['tweet', 'timestamp', 'followers', 'retweets','url count','user mention','favorited num','ranking score','hashtag count']
    date,time = [],[]
    for timestamp in data['timestamp']:
        pst_tz = pytz.timezone('US/Pacific')
        timestamp = str(datetime.datetime.fromtimestamp(int(timestamp), pst_tz))
        date_split = timestamp[0:10].split('-')
        date.append(int(date_split[0]+date_split[1]+date_split[2]))
        time.append(int(timestamp[11:13]))
    data.drop('timestamp', 1, inplace = True)
    data.insert(1, 'date', date)
    data.insert(2, 'time', time)
    df = data.groupby(['date', 'time']).agg({'date' : np.max, 'time' : np.max, 'tweet' : np.sum, 'retweets' : np.sum, 'followers' : np.sum, 'url count':np.sum,'user mention':np.sum,'favorited num':np.sum,'ranking score' : np.sum,'hashtag count':np.sum})
    addrow = []
    for i in range(1,len(df.index)):  
        oldday = df.index[i-1][0]
        oldtime = int(df.index[i-1][1])
        newday = df.index[i][0]
        newtime = int(df.index[i][1])
        if (newtime < oldtime):
            newtime = newtime + 24
        hour_diff = newtime - oldtime
        while (hour_diff > 1):
            oldtime = oldtime + 1
            if (oldtime > 23):
                oldday = newday
                addrow.append({'tweet':0,'date':oldday,'time':oldtime-24,'followers':0,'retweets':0,'url count':0,'user mention':0,'favorited num':0,'ranking score':0,'hashtag count':0})
            else:
                addrow.append({'tweet':0,'date':oldday,'time':oldtime,'followers':0,'retweets':0,'url count':0,'user mention':0,'favorited num':0,'ranking score':0,'hashtag count':0})
            hour_diff = newtime - oldtime
    for row in addrow:
        data = data.append(row, ignore_index=True)
    df = data.groupby(['date', 'time']).agg({'date' : np.max, 'time' : np.max, 'tweet' : np.sum, 'retweets' : np.sum, 'followers' : np.sum, 'url count':np.sum,'user mention':np.sum,'favorited num':np.sum,'ranking score' : np.sum,'hashtag count':np.sum})
    df.to_csv('part1_4'+file[15:-4]+'.csv', index=False)
    display(df)
    return df
def calculate_rmse(prediction, real):
    return np.sqrt(((prediction - real) ** 2).mean())
```

```python
def regression_analysis(file, df):
    input_arr = []
    for index in df.index:
        input_arr.append(df.loc[index, 'time':'hashtag count'].values)
    input_arr.pop()
    input_arr = sm.add_constant(input_arr)
    output_arr = df.loc[df.index[1]:, 'tweet'].values
    model = sm.OLS(output_arr, input_arr)
    results = model.fit()
    output_predicted = results.predict(input_arr)
    rmse = calculate_rmse(output_predicted, output_arr)
    print ('RMSE of '+file[15:-4]+' for the linear regression model is: '+str(rmse))
    print (results.summary())
    return output_predicted

choosenfeatures = {'#gohawks':['followers','url count','ranking score'], 
                 '#gopatriots':['favorited num','ranking score','hashtag count'], 
                 '#nfl':['user mention','favorited num','hashtag count'], 
                 '#patriots':['user mention','ranking score','hashtag count'], 
                 '#sb49':['retweets','ranking score','hashtag count'], 
                 '#superbowl':['favorited num','ranking score','hashtag count']}
def plot_choosen_features(file, df, output_predicted):
    features = choosenfeatures[file]
    opt = []
    for i in range(3):
        cur_feature = []
        for index in df.index:
            cur_feature.append(df.loc[index, features[i]:features[i]].values)
        cur_feature.pop()
        opt.append(cur_feature)
    
    fig = plt.figure(figsize=(18,5))
    fig.suptitle(file, y=1.1, fontsize=20)
    for i in range(3):
        ax = plt.subplot(131+i)
        ax.set_title('Predictant vs '+features[i], fontsize=18)
        ax.scatter(opt[i], output_predicted, color='red')
        ax.set_xlabel(features[i], fontsize=14)
        ax.set_ylabel('predicted number of tweets', fontsize=14)
    plt.tight_layout()
    plt.show()
    print ()
```

```python
def regression_model(file):
    df = preprocess2(file)
    output_predicted = regression_analysis(file, df)
    plot_choosen_features(file[14:-4],df,output_predicted)

for file in newfiles2:
    regression_model(file)
    
```
